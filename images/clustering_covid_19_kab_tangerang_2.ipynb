```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
import dataframe_image as dfi
```

```python
# df.to_csv("data_pasien_covid_19.csv", index=False)
df = pd.read_csv('data_pasien_covid_19_kab_tangerang.csv')
```

```python
df
```

```python
table = df.dfi.export('table_data_covid.png')
# dfi.export(df, 'table_data_covid.png')
```

```python
X = df.iloc[:, 2:9].values
X
```

```python
# standarisasi dengan minmax
nilai_standarisasi = preprocessing.MinMaxScaler().fit_transform(X)
nilai_standarisasi
```

```python
# menggunkan metode elbow untuk menentukan angka cluster yang tepat
# masukan kmeans
# within-cluster sum of squares atau wccs
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state=42)
  kmeans.fit(nilai_standarisasi)
  # print(kmeans)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah cluster')
plt.ylabel('WCSS')
gambar_metode_elbow = plt.gcf()
```

```python
gambar_metode_elbow.savefig('data_elbow.png', dpi=200)
```

```python
# Menjalankan K-Means clustering ke dataset
kmeans = KMeans(n_clusters=3, init = 'k-means++', random_state=123)
y_kmeans = kmeans.fit_predict(nilai_standarisasi)
print(kmeans)
print(y_kmeans)
```

```python
#visualisasi data 
# Visualisasi hasil clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 200, c = 'red', label = 'Cluster 0')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 200, c = 'orange', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 200, c = 'yellow', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'blue', label = 'Centroids')
plt.title('Clusters Persebaran Data Covid-19 di Kabupaten Tangerang')
plt.xlabel('Index cluster')
plt.ylabel('Rating  (1-100)')
plt.legend()
gambar_hasil_cluster = plt.gcf()
gambar_hasil_cluster.savefig('hasil_cluster.png', dpi=200)
plt.show()
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

def highlight_min(s):
    is_min = s == s.min()
    return ['background-color: red' if v else '' for v in is_min]

def highlight_else(s):
    is_min = s == s.min()
    is_max = s == s.max()
    is_else = ((is_min == False) & (is_max == False) )
    
    return ['background-color: orange' if v else '' for v in is_else]
dfs = df.style.apply(highlight_max, subset=['kluster']).\
apply(highlight_min, subset=['kluster']).\
apply(highlight_else, subset=['kluster'])
```
