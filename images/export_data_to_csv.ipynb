```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
df = pd.read_html('https://covid19.tangerangkab.go.id/sebaran-data')
df = df[0]
df.columns = df.columns.droplevel(1)
df.set_index('No', drop=True, inplace=True)
df.loc[:, ['Konfirmasi Sembuh', 'Konfirmasi Total', 'Jumlah  (Suspek Dirawat + Konfirmasi Total)']]  = df.loc[:, ['Konfirmasi Sembuh', 'Konfirmasi Total', 'Jumlah  (Suspek Dirawat + Konfirmasi Total)']].astype('str')
df = df.applymap(lambda x: '0' if x == '-' else x)
df.iloc[:, 2:] = df.iloc[:, 2:].apply(lambda x: x.astype('int64'))
df.info()
```
